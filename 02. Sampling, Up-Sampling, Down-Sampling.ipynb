```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Sampling
sampling_frequency=4000
amplitude=5
frequency=500
phase=90
phase = phase * np.pi /180
start = 0
end = .010
sample_count = sampling_frequency * (end - start)
x = np.linspace(start, end, int(sample_count), endpoint=False)
y = amplitude * np.sin((2* np.pi * frequency * x) + phase)
plt.stem(x,y)
plt.title("Sampling:")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.show()
    
```

```python
#Shifting

shift = 3
shifted_y = np.concatenate((np.zeros(shift), y[:-shift]))
plt.stem(x,shifted_y)
plt.title("Right shifted function:")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.show()
```

```python
#Up sampling
factor = 2;
up_sampled_y = np.zeros(np.size(y))

for i,j in zip(range(0,np.size(y)),range(0,np.size(up_sampled_y),2)):
    up_sampled_y[j] = y[i]
    
plt.stem(x,up_sampled_y[:np.size(y)])
plt.title("Up sampled signal:")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.show()
```

```python
#Down sampling
factor = 2;
down_sampled_y = y[::factor]


plt.stem(x[::factor],down_sampled_y)
plt.title("Down sampled signal:")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.show()
```
