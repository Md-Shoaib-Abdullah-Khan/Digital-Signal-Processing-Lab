```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Sampling
sampling_frequency=1000
amplitude=5
frequency=500
phase=90
phase = phase * np.pi/180
start = 0
end = .010
sample_count = sampling_frequency * (end - start)
x = np.linspace(start, end, int(sample_count), endpoint=False)
y = amplitude * np.sin((2* np.pi * frequency * x) + phase)
y=np.round(y,3)
plt.stem(x,y)
plt.title("Sampling:")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.show()
y
```

```python
#Shifting

shift = 3
shifted_y = np.concatenate((np.zeros(shift), y[:-shift]))
plt.stem(shifted_y)
plt.title("Right shifted function:")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.show()
```

```python
#Up sampling
factor = 2;
up_sampled_y = np.zeros((factor*np.size(y)))

up_sampled_y[::factor] = y 

t_upsampled = np.arange(len(up_sampled_y)) / (sampling_frequency * factor)

for i in range(1, len(up_sampled_y)-1, 2):
    up_sampled_y[i] = (up_sampled_y[i-1] + up_sampled_y[i+1])/2
    
plt.stem(t_upsampled, up_sampled_y)
plt.title("Up sampled signal:")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.show()
```

```python
#Down sampling
factor = 2;
down_sampled_y = y[::factor]


plt.stem(x[::factor],down_sampled_y)
plt.title("Down sampled signal:")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.show()
```
