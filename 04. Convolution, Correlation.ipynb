```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolve(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = np.zeros(len_y)
    for i in range(len_y):
        for k in range(len_x):
            if(i - k >= 0 and i - k < len_h):
                y[i] += x[k] * h[i - k]
    return y
```

```python
x = [2, 4, 6]
h = [3, 5, 7]

#Inverse h(x) for correlation
h_folded = h[::-1]

y = convolve(x, h)
y_corr = convolve(x, h_folded)


Nx = np.size(x) #Size of x
Nh = np.size(h) #Size of h
x_low = 0 #Lowest index of x
h_low = -1 #Lowest index of h

con_0 = x_low + h_low #Lowest index of convolution (x(n)*h(-n))

corr_0 = x_low - (h_low + Nh -1) #Lowest index of convolution (x(n)*h(-n))


plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.stem(range(x_low, x_low + len(x)), x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n)')
plt.axhline(color='r')
plt.axvline(color='r')
plt.grid()

plt.subplot(2, 2, 2)
plt.stem(range(h_low, h_low + len(h)), h)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('h(n)')
plt.axhline(color='r')
plt.axvline(color='r')
plt.grid()

plt.subplot(2, 2, 3)
plt.stem(range(con_0, con_0 + len(y)), y)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Convolution: y(n) = x(n) * h(n)')
plt.axhline(color='r')
plt.axvline(color='r')
plt.grid()

plt.subplot(2, 2, 4)
plt.stem(range(corr_0, corr_0 + len(y_corr)), y_corr)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Correlation: y(n) = x(n) * h(-n)')
plt.axhline(color='r')
plt.axvline(color='r')
plt.grid()


plt.tight_layout()
plt.show()
y
```
