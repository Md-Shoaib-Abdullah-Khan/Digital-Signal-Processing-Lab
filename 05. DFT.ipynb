```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def angl(real, img):
    if img == 0:
        return 0
    elif real == 0:
        if img > 0:
            return 90
        else:
            return -90
    else:
        return math.atan(img/real)*(180/np.pi)
    
```

```python
def dft(x):
    N = x.size
    X = np.zeros(N, dtype=complex)

    for m in range(0,N):
        for n in range(0,N):
            X[m] = X[m] + x[n] * np.exp((-2j * np.pi * n * m)/N)

    X = np.round(X,3)
    magnitude = np.abs(X)
 
    for i in range(N):
        phase[i] = angl(X[i].real, X[i].imag)
        if phase[i] > 90:
            phase[i] -= 90
        elif phase[i] < -90:
            phase[i] += 90
    

    return X, magnitude, phase
```

```python
x = np.array([.3535, .3535, .6464, 1.0607, .3535, -1.0607, -1.3535, -.3535])
X, magnitude, phase = dft(x)
N = np.size(x)

plt.figure(figsize=(10,12))
plt.subplot(3,3,1)
plt.plot(np.arange(N), x)
plt.title("x[n]")
plt.ylabel("Amplitude")
plt.xlabel("Frequency")
plt.grid()

plt.subplot(3,3,2)
plt.stem(np.arange(N), X.real)
plt.title("Real Value")
plt.ylabel("Amplitude")
plt.xlabel("Frequency")
plt.grid()

plt.subplot(3,3,3)
plt.stem(np.arange(N), X.imag)
plt.title("Imaginary Value")
plt.ylabel("Amplitude")
plt.xlabel("Frequency")
plt.grid()

plt.subplot(3,3,4)
plt.stem(np.arange(N), magnitude)
plt.title("Magnitude Spectrum")
plt.ylabel("Amplitude")
plt.xlabel("Frequency")
plt.grid()

plt.subplot(3,3,5)
plt.stem(np.arange(N), phase)
plt.title("Phase Spectrum")
plt.ylabel("Amplitude")
plt.xlabel("Frequency")
plt.grid()

plt.tight_layout()
plt.show()
```
