```python
def quantization(signal, L):
    
    step_size = (np.max(signal) - np.min(signal))/L
    levels = np.linspace(np.min(signal) + step_size/2, np.max(signal)-step_size/2, L)

    quantized_signal = np.zeros(len(signal))
    for i in range(len(signal)):
        quantized_signal[i] = levels[np.argmin(np.abs(signal[i]-levels))]

    return quantized_signal

    
```

```python
sampling_rate = 1000  # Samples per second
t = np.linspace(0, 1, sampling_rate)  # 1 second duration
original_signal = np.sin(2 * np.pi * 5 * t)  # Sine wave with 5 Hz frequency

L=8
quantized_signal = quantization(original_signal, L)

plt.figure(figsize=(10, 6))
plt.plot(t, original_signal, color="blue")
plt.step(t, quantized_signal, color="red", where="mid")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Quantization of a Signal in L Levels")
plt.grid(True)
plt.show()
```
