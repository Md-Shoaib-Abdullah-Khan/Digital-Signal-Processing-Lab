```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def unit_step():
    start = -10
    end = 10
    points = 20
    y = []
    x = np.linspace(start, end, points, endpoint=False)
    for i in range(20):
        if i<10:
            y.append(0)
        else:
            y.append(1)
    plt.stem(x,y)
    plt.title("Unit step function:")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.show()
```

```python
unit_step()
```

```python
def unit_ramp():
    start = -10
    end = 10
    points = 20
    y = []
    x = np.linspace(start, end, points, endpoint=False)
    for i in range(20):
        if i<=10:
            y.append(0)
        else:
            y.append(y[i-1] + 1)
    plt.stem(x,y)
    plt.title("Unit ramp function:")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.show()
```

```python
unit_ramp()
```

```python
def exponential():
    start = -5
    end = 5
    points = 20
    y = []
    x = np.linspace(start, end, points, endpoint=False)
    y = np.exp(x)
    plt.stem(x,y)
    plt.title("Exponential function:")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.show()
```

```python
exponential()
```

```python
def sine(amplitude, frequency, phase):
    phase = phase * np.pi /180
    start = -1
    end = 1
    points = 1000
    x = np.linspace(start, end, points, endpoint=False)
    y = amplitude * np.sin(2* np.pi * frequency * x + phase)
    plt.plot(x,y)
    plt.title("Sine function:")
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.show()
```

```python
sine(10, 1, 0)
```

```python
def cosine(amplitude, frequency, phase):
    phase = phase * np.pi /180
    start = -1
    end = 1
    points = 100
    x = np.linspace(start, end, points, endpoint=False)
    y = amplitude * np.cos(2* np.pi * frequency * x + phase)
    return x,y
```

```python
x,y = cosine(10, 1, 0)

plt.plot(x,y)
plt.title("Cosine function:")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
plt.show()
```
