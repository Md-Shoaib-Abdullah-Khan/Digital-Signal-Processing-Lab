```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolve(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = [0] * len_y
    for i in range(len_y):
        for k in range(len_x):
            if(i - k >= 0 and i - k < len_h):
                y[i] += x[k] * h[i - k]
    return y
```

```python
x = [1, 2, 3, 4, 5]
h = [2, 3, 4]
h_folded = h[::-1]
y = convolve(x, h)
y_corr = convolve(x, h_folded)

plt.subplot(2, 3, 1)
plt.stem(range(len(x)), x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n)')
plt.grid()

plt.subplot(2, 3, 2)
plt.stem(range(len(h)), h)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('h(n)')
plt.grid()

plt.subplot(2, 3, 3)
plt.stem(range(len(y)), y)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('y(n) = x(n) * h(n)')

plt.subplot(2, 3, 4)
plt.stem(range(len(x)), x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n)')
plt.grid()

plt.subplot(2, 3, 5)
plt.stem(range(len(h)), h)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('h(n)')
plt.grid()

plt.subplot(2, 3, 6)
plt.stem(range(len(y_corr)), y_corr)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('y(n) = x(n) . h(n)')
plt.tight_layout()
plt.show()
# print(y)
```

```python
h = np.array((1, 2, 1, 2))
x = np.array((3, 2, 1, 2))

n1 = 0
n2 = -1

transformed_h = np.flip(h,axis=0)

lh = len(h)
lx = len(x)

lh_zeros = np.zeros(lh-1);
lx_zeros = np.zeros(lx-1);

transformed_x = np.concatenate((lh_zeros, x))
transformed_h = np.concatenate((transformed_h, lx_zeros))

y = np.array([])

while len(transformed_x)>0 :
    sum = np.sum(transformed_x * transformed_h)
    transformed_x = np.delete(transformed_x, [0])
    transformed_h = np.delete(transformed_h, [len(transformed_h)-1])
    y = np.append(y,sum)
y

plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
plt.stem(range(-1,lx-1), x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n)')


plt.subplot(1,3,2)
plt.stem(range(0,lh), h)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('h(n)')

plt.subplot(1,3,3)
plt.stem(range(n1 + n2, lx+lh-2), y)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('y(n) = x(n) * h(n)')

plt.show()
```
